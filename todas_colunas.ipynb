```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:,.2f}'.format
%matplotlib inline

from sklearn.preprocessing import LabelEncoder

# Exemplo de uso do LabelEncoder
label_encoder = LabelEncoder()
```

```python
DS_treino = pd.read_csv("train.csv")
DS_teste = pd.read_csv("test.csv")

DS_treino = DS_treino.fillna(0)
DS_teste = DS_teste.fillna(0)
```

```python
colunas = [
    'MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 
    'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 
    'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 
    'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
    'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 
    'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 
    'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition'
]

# Para cada coluna, primeiro converte para string e depois aplica o LabelEncoder
for coluna in colunas:
    DS_treino[coluna] = DS_treino[coluna].astype(str)  # Converte para string
    DS_treino[coluna] = label_encoder.fit_transform(DS_treino[coluna])
    DS_teste[coluna] = DS_teste[coluna].astype(str)  # Converte para string
    DS_teste[coluna] = label_encoder.fit_transform(DS_teste[coluna])
```

```python
Y_treino = DS_treino.SalePrice
X_treino = DS_treino.drop('SalePrice', axis=1)

X_teste = DS_teste
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
modelo_regressao = LinearRegression()

modelo_regressao.fit(X_treino, Y_treino)

Y_regressao = modelo_regressao.predict(X_teste)
data = {'ID':X_teste['Id'], 'predição' : Y_regressao}
data = pd.DataFrame(data)
data
```

```python
arvore = DecisionTreeRegressor()

arvore.fit(X_treino, Y_treino)

Y_regressao = arvore.predict(X_teste)
data = {'ID':X_teste['Id'], 'predição' : Y_regressao}
data = pd.DataFrame(data)
data
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(X_treino, Y_treino)

modelo_regressao.fit(x_treino, y_treino)

Y_regressao = modelo_regressao.predict(x_teste)

DS_final = pd.DataFrame({'teste': y_teste, 'Predição': Y_regressao})
DS_final = DS_final.sort_values(by='teste', ascending=False)
DS_final
```

```python
arvore.fit(x_treino, y_treino)

Y_regressao = arvore.predict(x_teste)

DS_final = pd.DataFrame({'teste': y_teste, 'Predição': Y_regressao})
DS_final = DS_final.sort_values(by='teste', ascending=False)
DS_final.head(10)
```
